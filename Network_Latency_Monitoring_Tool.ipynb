Latency: The delay between sending a request and receiving a response.
RTT: Time taken for a data packet to go to the destination and back.
Ping: A tool that sends ICMP Echo Requests to a host and waits for a reply to measure RTT.

```python
from ping3 import ping

# Enter the IP or domain name of the host
host = "google.com"

# Send a ping and get the response time in seconds
response_time = ping(host)

if response_time:
    print(f"Latency: {response_time * 1000:.2f} ms")
else:
    print("Request timed out")
```

```python
from ping3 import ping
import time

host = "google.com"
ping_count = 10
latency_results = []

for i in range(ping_count):
    latency = ping(host)
    if latency:
        print(f"Ping {i+1}: {latency * 1000:.2f} ms")
        latency_results.append(latency * 1000)
    else:
        print(f"Ping {i+1}: Request timed out")
    time.sleep(1)  # Wait 1 second before next ping

# Calculate average latency
if latency_results:
    average_latency = sum(latency_results) / len(latency_results)
    print(f"\nAverage Latency: {average_latency:.2f} ms")
else:
    print("All requests timed out")
```

```python
from ping3 import ping
import matplotlib.pyplot as plt
import time

host = "google.com"
ping_count = 10
latency_results = []

for i in range(ping_count):
    latency = ping(host)
    if latency:
        latency_results.append(latency * 1000)
    else:
        latency_results.append(None)
    time.sleep(1)  # Wait 1 second before next ping

# Plot the latency results
plt.plot(latency_results, marker='o')
plt.title(f"Network Latency to {host}")
plt.xlabel("Ping Attempt")
plt.ylabel("Latency (ms)")
plt.grid(True)
plt.show()
```

```python
try:
    latency = ping(host, timeout=2)  # 2-second timeout
except Exception as e:
    print(f"Error: {e}")
```
